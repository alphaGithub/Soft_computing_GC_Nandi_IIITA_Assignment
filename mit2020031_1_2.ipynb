## Implementing Box-Muller Transformation algorithm in Python (useNumPy library only).

In the Probability and Satistics course (PBS-432), you perhaphs have implemented the concept of Box-Muller transformation. If not, please learn about it now, because it is very useful and has wide range of applications. In the machine learning Gaussian distribution/the Normal Distribution is the workhorse and being able to draw samples from this distribution lies at the heart of many machine learning algorithms. Box-Muller transform enables us to transform any i.i.d (independently identical distribution), say uniform distribution U(0,1) to i.i.d normal distribution N(0,1).


```python
#Loading required package
import numpy as np
import matplotlib.pyplot as plt       
```

```python
#u1,u2 random variable between 0 to 1 with uniform distribution
u1,u2 = np.random.uniform(0,1,4096),np.random.uniform(0,1,4096)
plt.hist(u1,bins=32)
plt.show()
plt.hist(u2,bins=32)
plt.show()
```

```python
#using Box Muller transformation
theta = 2*3.1415*u2
sqrt_log_u1 = np.sqrt(-2*np.log(u1))
x,y = sqrt_log_u1*(np.cos(theta),np.sin(theta))
```

### forumula use to transform u1,u2 to x,y(i.e uniform distribution to normal distribution)

x = (-2ln(u1)^1/2  * cos(2*3.141*u2)

y = (-2ln(u1)^1/2  * sin(2*3.141*u2)

u1,u2 are uniform distribution and x,y are normal distribution

```python
#Plotting x,y independent random variable with Normal distribution
plt.hist(x,bins=32)
plt.show()
plt.hist(x,bins=32)
plt.show()
```
