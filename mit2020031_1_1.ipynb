# Mean and Varience

Problem statement :\
Create a data set of 200 entries (could be anything age, marks, height, weight
etc.). To get unbiased estimation, using a random function generator, collect
15 set samples of data of 10 classes having 20 entries each. Calculate the
mean and variance of each sample classes as well as the population (of 200).
Note down your observations.

```python
#loading required package
import numpy as np
import random
```

generateClass(class_range) : generate classes for each field depending on range \
    for example: \
    field = age \
    range = 18-60 \
    data will be generated for 20 data items between 18-60 \
    This will be repeated for every field to generate class

generateSample():  generates sample containing 10 class each having 20 entries in each field(age,height,wieght,...)


```python
 def generateClass(class_range):
    age = np.random.randint(class_range[0][0],class_range[0][1],20)
    height = np.random.randint(class_range[1][0],class_range[1][1],20)
    weight = np.random.randint(class_range[2][0],class_range[2][1],20)
    iq = np.random.randint(class_range[3][0],class_range[3][1],20)
    salary_per_day = np.random.randint(class_range[4][0],class_range[4][1],20)
    sleep_hour = np.random.randint(class_range[5][0],class_range[5][1],20)
    marks = np.random.randint(class_range[6][0],class_range[6][1],20)
    avg_distance = np.random.randint(class_range[7][0],class_range[7][1],20)
    sugar_level = np.random.randint(class_range[8][0],class_range[8][1],20)
    blood_pressure = np.random.randint(class_range[9][0],class_range[9][1],20)
    return [age,height,weight,iq,salary_per_day,sleep_hour,marks,avg_distance,sugar_level,blood_pressure]
    
def generateSample():
    cr1 = [(18,60,),(150,190),(50,120),(70,120),(500,1500),(4,9),(60,100),(1,5),(80,200),(60,150)]
    cr2 = [(18,25,),(150,190),(60,100),(90,120),(300,1000),(4,8),(65,100),(1,9),(80,140),(70,140)]
    cr3 = [(25,50,),(150,190),(50,130),(80,120),(600,2400),(4,7),(60,100),(1,8),(80,200),(60,150)]
    cr4 = [(50,60,),(140,180),(50,110),(70,100),(500,1700),(6,9),(50,85),(1,5),(60,200),(55,150)]
    cr5 = [(25,30,),(160,190),(60,120),(80,130),(400,2000),(4,7),(75,100),(1,8),(80,170),(75,140)]
    cr6 = [(30,60,),(150,190),(60,120),(80,120),(600,2500),(4,9),(70,100),(1,5),(80,210),(60,150)]
    cr7 = [(5,20, ),(130,190),(20,80),(70,120),(1,800),(6,10),(40,90),(1,4),(80,200),(80,140)]
    cr8 = [(20,40,),(155,190),(50,120),(80,120),(500,1200),(4,8),(60,100),(1,8),(80,200),(70,140)]
    cr9 = [(40,60,),(145,180),(50,120),(70,100),(200,2000),(6,9),(40,90),(1,6),(90,200),(60,145)]
    cr10 = [(60,75,),(130,170),(50,120),(70,90),(100,300),(7,9),(40,80),(1,4),(50,210),(50,150)]
    class_range = [cr1,cr2,cr3,cr4,cr5,cr6,cr7,cr8,cr9,cr10]
    sample = []
    for cr in class_range:
        c = generateClass(cr)
        sample.append(c)
    return sample
generateSample() #generating a sample
```

```python
# function to compute mean of data
# data = [d1,d2,....,dn]
def mean(data):
    sum_of_data = 0
    l=0
    for x in data:
        sum_of_data += x
        l += 1
    return sum_of_data/l
```

```python
#function to compute varience of data
#data = [d1,d2,.....,dn]
#mean_data = mean of data
def varience(data,mean_data):
    s = 0
    n = 0
    for x in list(data):
        s += (x-mean_data)**2
        n += 1
    return s/(n-1)
```

```python
#Generating all 15 sample one by one
sample_set = []
for i in range(0,15):
    sample = generateSample()
    sample_set.append(sample)
    
```

```python
#for sample 1
#showing mean of each class and population
np.set_printoptions(suppress=True)      # for suppression of scientific notation while printing


i=1
mean_all = []
var_all = []
for class_i in sample_set[0]:
    mean_class=[]
    var_class=[]
    for field in class_i:
        mean_class.append(mean(field))
        var_class.append(varience(field,mean(field)))
        
    print("class = ",i)
    print(mean_class)
    print(list(np.around(var_class,2)))
    i += 1


for i in range(10):
    data_field = []
    for class_i in sample_set[0]:
        data_field += list(class_i[i])
    mean_all.append(mean(data_field))
    var_all.append(varience(data_field,mean(data_field)))
print("\n----------------------Population----------------------------")
print("Mean\n",mean_all)
print("Varience\n",list(np.around(var_all,2)))
```
